```python
import matplotlib.pyplot as plt

degs = [35, 36, 45, 46, 55, 56]

for deg in degs:
    for data in [f"{deg}shell.dat"]:
        x = []
        y = []
        with open(data) as f:
            for l in f:
                _, xx, yy, _ = l.split(' ')
                x.append(float(xx)/1000)
                y.append(float(yy)/1000)
            plt.plot(x, y)

plt.title('Projectile trajectory')
plt.xlabel('x (km)')
plt.ylabel('y (km)')
plt.show()
```
